```r

df = read.csv("C:/Users/student/Desktop/Senior/MATH 421 - R/assignment/train.csv")
#correct data type
df$Survived = factor(df$Survived)
df$Pclass = factor(df$Pclass)
df$Name = NULL
df$PassengerId = NULL
df$Ticket = NULL
df$Cabin = NULL
df$NumPeople = df$SibSp + df$Parch  #combine sibsp and parch as NumPeople
df$SibSp = NULL
df$Parch = NULL
str(df)


checkmiss = function(data)
{
  for (i in 1:ncol(data)){
      num = sum(is.na(data[,i]))  
      print(num)
  }
} 
checkmiss(df)


varmiss = function(data)
{
  for (i in 1:ncol(data)){
    num = sum(is.na(data[, i]))
    if (num != 0 ){
      print(names(data[i]))
    }
  }  
}
varmiss(df)

combine= function(data){
  for (i in 1:ncol(data)){
    num = sum(is.na(data[,i]))
    if (num != 0 ){
      print(c(names(data)[i], num))
    }
  }
}
combine(df)

handlemiss = function(data)
{
  for (i in 1:ncol(data)){
    if (class(data[, i]) != "factor") 
    { 
      data[,i][is.na(data[,i])]= mean(data[,i], na.rm = TRUE)   
    } else {
      levels=unique(data[,i])
      data[,i][is.na(data[,i])]=levels[which.max(tabulate(match(data[,i], levels)))]
    } 
  }
  return(data)
}
newdata = handlemiss(df)
sum(is.na(newdata$Age)) 

splitdata=function(data){
  subnum=c()
  subcat=c()
  for (i in 1:ncol(data)){
    if(class(data[, i]) != "factor"){
      subnum=c(subnum,names(data)[i])
    }else{
      subcat=c(subcat,names(data)[i])
    }
  }
  check = head(subnum)
  return(check)
}
splitdata(newdata)


library(ggplot2)
dencurve = function(data){
  for (i in 1:ncol(data)){
    if (class(data[,i])!="factor") {
    curve = ggplot(data) + geom_density(mapping = aes(x=data[,i]), na.rm = TRUE) +labs( x = names(data)[i])
    print(curve)
    }
  }
}
dencurve(newdata)


bargraph = function(data){
  for (i in 1:ncol(data)){
    if (class(data[,i])=="factor") {
    curve = ggplot(data) + geom_bar(mapping = aes(x=data[,i])) + xlab(names(data[i]))
    print(curve)
  }
  }
}
bargraph(newdata)



two_var_bargraph = function(data, k, j){
  
    if (class(data[,k])!="factor"|class(data[,j])!= "factor") {
      print("Please input two categorical variables")
    }
      else {
        curve = ggplot(data) + geom_bar(mapping = aes(x=data[, k], fill = data[, j]), position  = "dodge") + labs(x = names(data[j]), fill = names(data[j]))
        print(curve)
      }
  }
two_var_bargraph(newdata, 1, 2)


two_var_bargraph2 = function(data, k, j){
  
    if (class(data[,k])=="factor"|class(data[,j])== "factor") {
      print("Please input two numeric variables")
    }
      else {
        curve = ggplot(data) + geom_point(mapping = aes(x=data[, k], y = data[, j])) + labs(x = names(data[k]), y = names(data[j]))
        print(curve)
      }
  }
two_var_bargraph2(newdata, 4, 5)


two_var_bargraph3 = function(data, k, j){
  
    if (class(data[,k])=="factor") {
      print("Please input numeric variable")
    }
      else if (class(data[, j]) != 'factor'){
        print("Partion variable not categorical")
  }
  else {curve = ggplot(data) + geom_density(mapping = aes(x=data[, k], color = data[, j])) + labs(x = names(data[k]), color = names(data[j]))
        print(curve)
      }
  }
two_var_bargraph3(newdata, 4, 1)


bar=function(data){
    for (i in 1:ncol(data)){
      if (!is.numeric(data[,i])){   #is.numeric returns the logic is a variable is numeric
        for(j in 1:ncol(data)){
          if (!is.numeric(data[,j]) & j!=i){
            curve = ggplot(data)+geom_bar(mapping=aes(x=data[,i],fill=data[,j]),position="dodge")+labs(x=names(data)[i],fill=names(data)[j])
            print(curve)
          }
        }
      }
    }
}


bar(newdata)


den =function(data){
  for (i in 1:ncol(data)){
    if(is.numeric(data[, i])){  #if numeric the variable will have a density funciton by itself
        curve = ggplot(data) + geom_density(mapping = aes(x=data[,i]), na.rm = TRUE) +labs( x = names(data)[i])
        print(curve)
      for (j in 1:ncol(data)){  #or a density curve can be a combination of numeric and categorical
        if(!is.numeric(data[, j])){
          curve = ggplot(data) + geom_density(mapping = aes(x = data[, i], color = data [ , j])) + labs(x = names(data)[i], color = names(data)[j]) 
          print(curve)

        }
      }
    }
  }
}


den(newdata)


graphall = function(data){
  for (i in 1:ncol(data)){
    if(is.numeric(data[, i])){  #if numeric the variable will have a density funciton by itself
        curve = ggplot(data) + geom_density(mapping = aes(x=data[,i]), na.rm = TRUE) +labs( x = names(data)[i])
        print(curve)
      for (j in 1:ncol(data)){  #or a density curve can be a combination of numeric and categorical
        if(!is.numeric(data[, j])){
          curve = ggplot(data) + geom_density(mapping = aes(x = data[, i], color = data [ , j])) + labs(x = names(data)[i], color = names(data)[j]) 
          print(curve)

        }
      }
    }
    else #if not numeric then should have bar chart - apply bar function 
    {
       bar(data)
    } 
  }
}
graphall(newdata)



```
